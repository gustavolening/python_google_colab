#<font color = 'Yellow'> Manipulando os dados

## <font color = 'Green'>Criando colunas numéricas

```python
url = ('https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv')
dados = pd.read_csv(url, sep=";")
dados.head()
#Importando os dados novamente para pegar a base do 0.
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
#Criando coluna Valor_por_mes que é a soma da coluna Valor + Condominio
```

```python
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU']
dados.head()
#Criando a coluna Valor_por_ano que é a coluna Valor_por_mes * 12 (quantidade de meses doa no) + IPTU (que é cobrado 1 vez por ano)
```

## <font color = 'Green'>Criando colunas categóricas

```python
from pandas.core.internals import concat
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro']
dados.head()
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + dados['Quartos'].astype(str) + ' quarto(s) ' + ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garem.'
dados.head()
#Criando coluna de descrição mesclando Tipo, Bairro, Quartos e Vagas e convertendo Quartos e Vagas para string.
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x >0 else "Nao")
dados.head()
#Criando coluna de True e False com o lambda para usar condicional if
```

```python
dados.to_csv('dados_completos_dev.csv', index=False, sep=';')
```
