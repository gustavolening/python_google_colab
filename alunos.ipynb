```python
df = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv')
df.head()
```

###**<font color = 'yellow'> Os alunos participaram de uma atividade extracurricular e ganharam pontos extras. Esses pontos extras correspondem a 40% da nota atual de cada um deles. Com base nisso, crie uma coluna chamada "Pontos_extras" que contenha os pontos extras de cada aluno, ou seja, 40% da nota atual deles.**

```python
df['Notas_finais'] = df['Notas'] + df['Pontos_extras']
df.head()
```

###**<font color = 'yellow'> Crie mais uma coluna, chamada "Notas_finais" que possua as notas de cada aluno somada com os pontos extras.**

```python
df['Aprovado_final'] = df['Notas_finais'].apply(lambda x: "Aprovado" if x >= 6 else "Reprovado")
df.head()
```

###**<font color = 'yellow'> Como houve uma pontuação extra, alguns alunos que não tinham sido aprovados antes podem ter sido aprovados agora. Com base nisso, crie uma coluna chamada "Aprovado_final" com os seguintes valores:**
*True: caso o aluno esteja aprovado (nota final deve ser maior ou igual a 6);*

*False: caso o aluno esteja reprovado (nota final deve ser menor que 6).*

```python
df.query('Aprovado == False & Aprovado_final == "Aprovado"')
```

###**<font color = 'yellow'>Faça uma seleção e verifique quais alunos não tinham sido aprovados anteriormente, mas foram aprovados após a soma dos pontos extras.**

```python
selecao = (df['Aprovado'] == False) & (df['Aprovado_final'] == 'Aprovado')
df[selecao]
```
